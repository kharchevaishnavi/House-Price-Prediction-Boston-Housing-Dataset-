Problem Statement:
Predict the median value of owner-occupied homes (MEDV) based on various features such as crime rate, number of rooms, property tax, etc.

Description:Train a regression model to predict house prices using the classic Boston Housing dataset or a similar dataset from Kaggle.
Challenges:
Feature scaling and encoding categorical data
Train/test split and evaluate model using RMSE or R²
Visualize model performance and key features
Tech Stack: Python, Pandas, NumPy, Scikit-learn, Matplotlib

1.Import Required Libraries

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

2.Load Dataset

```python
df = pd.read_csv("BostonHousing.csv")
```

3.Exploratory Data Analysis

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

4.Split Features and Target

```python

X = df.drop('medv', axis=1)
y = df['medv']
```

5.Feauture Scaling

```python

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

6.Train Test Split

```python

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

7. Train Regression Model

```python

model = LinearRegression()
model.fit(X_train, y_train)
```

8.Make predictions

```python
y_pred = model.predict(X_test)
```

9.Model Evalution

```python
##RMSE

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE:",rmse)
```

```python
##R2 Score

r2 = r2_score(y_test,y_pred)
print("R2 Score:",r2)
```

10.Visualization

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted House Prices")
plt.show()
```
